### 1.

Marginal probability:
\begin{align}
\pi(a) & \propto \int_0^{\infty} a^4 b^6 e^{-a-b-3ab} db \\
        &=  a^4 e^{-a} \int_0^{\infty} b^6 e^{-b-3ab} db\\
        &= \frac{\Gamma(7) a^4 e^{-a}}{(3a+1)^7}
\end{align}

\begin{align}
\pi(b) & \propto \int_0^{\infty} a^4 b^6 e^{-a-b-3ab} da \\
        & = b^6 e^{-b} \int_0^{\infty} a^4 e^{-a-3ab} da \\
        & =  \frac{\Gamma(5) b^6 e^{-b}}{(3b+1)^5}
\end{align}

Conditional probablity:
\begin{align}
\pi(a|b) &= \frac{\pi (a,b)}{\pi(b)} \\
        & \propto \frac{a^4 b^6 e^{-a-b-3ab} (3b+1)^5}{\Gamma(5) b^6 e^{-b}} \\
        & = \frac{a^4e^{-a-3ab} (3b+1)^5}{\Gamma(5)} \\
        & \propto a^4 e^{-a(1+3b)}
\end{align}

\begin{align}
\pi(b|a) &= \frac{\pi (a,b)}{\pi(a)} \\
        & \propto \frac{a^4 b^6 e^{-a-b-3ab} (3a+1)^7}{\Gamma(7) a^4 e^{-a}} \\
        & = \frac{b^6 e^{-b-3ab} (3a+1)^7}{\Gamma(7)} \\
        & \propto b^6 e^{-b(1+3a)}
\end{align}

Which turns out to be two gamma functions and we can easily sample from them.

```python
def draw_a_b(b):
    return np.random.gamma(7, 1/(1+3*b))

def draw_b_a(a):
    return np.random.gamma(5, 1/(1+3*a))
```

```python
a = np.random.gamma(2, 2)
N = 10000
sample = np.zeros((N, 2))
for i in range(N):
    b = draw_b_a(a)
    a = draw_a_b(b)
    sample[i, 0] = a
    sample[i, 1] = b
print "The MC integration is: ", sample[:,0].dot(sample[:,1])/N
```

To verify that this sampling is correct, the histograms is plotted. Histogram of $a$:

```python
sns.distplot(sample[:,0])
```

Marginal distribution of a:

```python
X = np.linspace(0, 14, 101)
plt.plot(X, X**4*np.exp(-X)/(3*a+1)**7)
```

Histogram of b:

```python
sns.distplot(sample[:,1])
```

Marginal distribution of b:

```python
plt.plot(X, X**6*np.exp(-X)/(3*X+1)**5)
```

### 2. Metropolis-Hasting


Stationary density is proportional to:

```python
def pi(X):
    return np.exp(-0.5*X**2)/(1+X**2)
```

Plot of this density function:

```python
X = np.linspace(-5, 5, 101)
plt.plot(X, pi(X))
```

Alpha for Metropolis-Hasting: since the propoal density is normal distribution, $q(X_2|X_1) =q(X_1|X_2) $

```python
def alpha(X2, X1):
    return min(1, pi(X2)/pi(X1))
```

Metropolis-Hasting sampling:

```python
def transit(X1, sigma):
    return np.random.normal(X1, sigma)
```

```python
def MetropolisHasting(sigma, N=1000):
    X1 = np.random.normal(0, 1) #q0 is standard normal distribution
    move_cnt = 0
    samples = np.zeros(N)
    for i in range(N):
        X2 = transit(X1, sigma) #proposal distribution
        u = np.random.uniform() #unifrom distribution between (0,1)
        if u < alpha(X2, X1):
            X2 = X2
            move_cnt += 1
        else:
            X2 = X1
        samples[i] = X2
        X1 = X2

    sns.distplot(samples)
    print "Number of moves: ", move_cnt
    return samples
```

i. When $\sigma$ is too small, the Markov-chain always moves, however not very far from the previous sample.

```python
sigma = 0.001
MetropolisHasting(sigma, 10000)
```

ii. when $\sigma$ is too big, the samples get rejected most of the time and barely move.

```python
sigma = 1000
MetropolisHasting(sigma, 10000)
```

Choose $\sigma = 2$, which gets an acceptance rate between 30% - 40%:

```python
sigma = 2
samples = MetropolisHasting(sigma, 100000)
```

We can see that the histogram is consistent with the actual density plot.

Use this sampling to estimate the value of the integral:

\begin{align}
I = \int x \pi(x) dx = \int \frac{x}{1+x^2} e^{-x^2/2} dx
\end{align}

The integrand is an odd function, thus the actual integral should be 0.

```python
I = samples.mean()
print "The MC approximated integral: ",I
```

If instead we use $q(x' | x) = N(x' | -x, \sigma^2)$:

```python
def transit(X1, sigma):
    return np.random.normal(-X1, sigma)

sigma = 2
samples = MetropolisHasting(sigma, 100000)
I = samples.mean()
print "The MC approximated integral: ",I
```

The simulation result is similar since the sationary density and proposal density are both even functions.

\begin{align}
\int_{-\infty}^{\infty} \pi(x) N(x' | x, \sigma^2) dx &= \int_{-\infty}^{\infty} \frac{1}{\sqrt{2\pi\sigma^2} (1+x^2)} Exp[\frac{(x'-x)^2}{2 \sigma^2} - \frac{x^2}{2}] dx \\
& = \int_{-\infty}^{\infty} \frac{1}{\sqrt{2\pi\sigma^2} (1+x^2)} Exp[\frac{(x'+(-x))^2}{2 \sigma^2} - \frac{(-x)^2}{2}] d(-x) \\
& = \int_{-\infty}^{\infty} \pi(x) N(x' | -x, \sigma^2) dx
\end{align}

### 3.

Use Metropolis-Hasting with a proposal density of $q(\theta_2 | \theta_1) = N(\theta_2 | \theta_1, 2)$

```python
def pi3(X):
    return np.exp(X) * np.exp(-np.exp(X)) * np.exp(-X**2/2)
```

```python
def alpha3(X2, X1):
    return min(1, pi3(X2)/pi3(X1))
```

```python
def MetropolisHasting3(sigma = 1, N = 1000):
    X1 = np.random.normal(0, 1) # q0 is standard normal distribution
    move_cnt = 0
    samples = np.zeros(N)
    for i in range(N):
        X2 = transit(X1, sigma) # proposal distribution
        u = np.random.uniform() # unifrom distribution between (0,1)
        if u < alpha3(X2, X1):  # use new alpha 
            X2 = X2
            move_cnt += 1
        else:
            X2 = X1
        samples[i] = X2
        X1 = X2

    sns.distplot(samples)
    print "Number of moves: ", move_cnt
    return samples
```

Draw 10000 samples and plot histogram:

```python
samples = MetropolisHasting3(2, 100000)
```

Plot the actual density function:

```python
X = np.linspace(-3.5, 2.5, 101)
plt.plot(X, pi3(X))
```
