```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc
#from __future__ import division
#rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
#rc('text', usetex=True)
%matplotlib inline
```

```python
x = 1
X = []
for i in range(50):
    x = np.random.normal(x, 1)
    X.append(x)
Y = [np.random.normal(x, 1) for x in X]
plt.plot(X, Y)
plt.plot(np.linspace(min(X), max(X), 100), np.linspace(min(X), max(X), 100), '--')
```

```python
sns.distplot(X, label='X')
sns.distplot(Y, label='Y')
plt.title('Data Distribution')
plt.legend()
plt.show()
```

## Kalman Filter

```python
mu = Y[0]
sigma2 = 1
sigma2_list = [sigma2]
mu_list = [mu]
for i in range(1, len(Y)):
    mu, sigma2 = (Y[i]*(1+sigma2) + mu)/(2+sigma2), (1+sigma2)/(2+sigma2)
    sigma2_list.append(sigma2)
    mu_list.append(mu)
```

```python
plt.plot(X, '--', label='X')
plt.plot(Y, color='y', label='Y')
plt.errorbar(range(50), mu_list, yerr=2*np.sqrt(sigma2_list), fmt='.', color='black', ecolor='lightgrey', elinewidth=3, label='95% Kalman Filter')
plt.legend()
plt.xlabel('iteration')
plt.show()
```

```python
Y = np.array(Y)
X = np.array(X)
mu_list = np.array(mu_list)
lambda_list = np.array(lambda_list)
```

```python
def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.))) / np.sqrt(2*np.pi) / sig
upper = mu + 5*np.sqrt(sigma2)
lower = mu - 5*np.sqrt(sigma2)
plt.plot(np.linspace(lower,upper,100),gaussian(np.linspace(lower,upper,100), mu, np.sqrt(sigma2)))
```

## Particle filter

The particle filter is repeated 1000 times to verify this implementation is correct.

```python
N = 1000
particles = np.ones(N)
weight = np.ones(N)/np.ones(N).sum()

threshold = 0.999
dist = []
for j in range(1000):
    particles = np.ones(N)
    weight = np.ones(N)/np.ones(N).sum()
    for i in range(50):
        particles = np.random.normal(particles, 1)
        weight = weight*gaussian(Y[i], particles, 1)
        weight /= weight.sum()
    dist += list(particles[weight>threshold])
print(len(dist))
```

### Take the weight-concentrated particles and plot histogram

```python
plt.plot(np.linspace(lower,upper,100),gaussian(np.linspace(lower,upper,100), 
                                               mu, np.sqrt(sigma2)), label = 'Kalman filter estimation')
sns.distplot(dist, label = 'single weighted particle')
plt.legend()
```
